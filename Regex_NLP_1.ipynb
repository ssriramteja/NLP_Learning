**1.** **re.findall(pattern, text)**

	â€¢	This searches through the text using the given regex pattern.

	â€¢	It returns all matches as a list.


**2. if matches:**

	â€¢	Checks if the list is not empty.

	â€¢	If at least one match exists â†’ True.


**3. return matches[0]**

	â€¢	Returns the first match from the list.

	â€¢	This is useful when you expect only one value (like an order number).


```python
def get_pattern_match(pattern, text):
    matches = re.findall(pattern, text)
    if matches:
        return matches[0]
```

```python
get_pattern_match(r'order[^\d]*(\d*)', chat1)
```

```python
def extract_personal_information(text):
    age = get_pattern_match(r'age (\d+)', text)
    full_name = get_pattern_match(r'Born(.*)\n', text)
    birth_date = get_pattern_match(r'Born.*\n(.*)\(age', text)
    birth_place = get_pattern_match(r'\(age.*\n(.*)', text)
    return {
        'age': int(age),
        'name': full_name.strip(),
        'birth_date': birth_date.strip(),
        'birth_place': birth_place.strip()
    }
```
